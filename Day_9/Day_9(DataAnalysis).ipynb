```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#using matplot
x=np.array([1,2,3])
y=np.array([1,5,3])
plt.plot(x,y)
```

```python
#using seaborn(lineplot)
x=np.array([1,2,3])
y=np.array([1,5,3])
sns.lineplot(x=x,y=y)
plt.show()
```

```python
#lineplot
x=np.array([1,2,3])
y=np.array([1,5,3])
df=pd.DataFrame({
     'x':x,
     'y':y
})
plt.figure(figsize=(2,2))
sns.lineplot(x=x,y=y,data=df)
plt.show()
```

```python
df= sns.load_dataset('tips')
#df
#df.head()
#df.head(10)
#print(df.to_string())
df.tail()
```

```python
#barplot
plt.figure(figsize=(3,40))
sns.barplot(x='time',y='total_bill',data=df,hue='sex',palette='coolwarm',orient='h')
plt.show()
```

```python
#using countplot
plt.figure(figsize=(3,3))
sns.countplot(x='time',data=df)
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(y='time',data=df)
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.lineplot(x='time',y='total_bill',data=df,hue='sex',style='sex',legend=False,dashes=False)
plt.show()
```

```python
#histogram
sns.distplot(df['total_bill'],kde=False,color='red')
plt.grid()
plt.show()
```

```python
#scatter plot
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex',style='sex',markers=['s','D'],palette='dark')
```

```python
#boxplot
sns.boxplot(y='total_bill',x='time',data=df)
plt.show()
```

```python
#violoneplot
sns.violinplot(y='total_bill',x='time',data=df,hue='sex',inner='quarter')
plt.show()
```

```python
#heatmap
#plt.figure(figsize=(2,2))
a=np.array([[1,20,30],[10,20,3],[1,2,3]])
sns.heatmap(a,cmap='PuOr')
plt.show()
```

```python
plt.figure(figsize=(2,2))
a=np.array([[1,20,30],[10,20,3],[1,2,3]])
b=[['a1','a2','a3'],['b1','b2','b3'],['c1','c2','c3']]
c={'fontsize':10,'color':'black'}
sns.heatmap(a,cmap='PuOr',annot=b,fmt='s',annot_kws=c,cbar=False,linewidth=4,linecolor='grey')
plt.show()
```
